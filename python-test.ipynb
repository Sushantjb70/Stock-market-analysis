```python
import pandas as pd
import matplotlib.pyplot as plt
from nsepy import get_history
from datetime import date
data = get_history(symbol="INFY", start=date(2015,1,1), end=date(2016,1,31))
data1= get_history(symbol="TCS",start=date(2015,1,1), end=date(2016,1,31))
```

```python
data
```

```python
data1
```

```python
data_close = pd.DataFrame(data.Close)
```

```python
data_close['28'] = data_close.Close.rolling(28).mean().shift()
data_close['56'] = data_close.Close.rolling(56).mean()
data_close['84'] = data_close.Close.rolling(84).mean()
data_close['112'] = data_close.Close.rolling(112).mean()
data_close['140'] = data_close.Close.rolling(140).mean()
```

```python
data_close['28'].head()
```

```python
data_close['140'].head()
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(data_close['Close'],label='Data')
plt.plot(data_close['28'], label='28 day')
plt.plot(data_close['56'], label='56 day')
plt.plot(data_close['84'], label='84 day')
plt.plot(data_close['112'], label='112 day')
plt.plot(data_close['140'], label='140 day')
plt.legend(loc=2)
```
